**Collect the internal energy (E) calculated by Quantum ESPRESSO for several lattice constants (a).**

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([5.0, 5.1, 5.2, 5.3, 5.43, 5.5, 5.6])
E = np.array([-15.79415749, -15.81972595, -15.83659006, -15.84608739, \
              -15.84906526, -15.84699148, -15.84029589])

# plot
fig, ax = plt.subplots()
ax.plot(a, E, 'o-', linewidth=2, markersize=10)

# figure/axes style (equiv. to set(gcf/gca, ...) + grid on, box on)
fig.patch.set_facecolor('white')
ax.tick_params(labelsize=20)
ax.grid(True)
ax.set_frame_on(True)  # 'box on'

# axis range (like MATLAB axis([...]))
rg = np.ptp(E)                      # max(E) - min(E)
tv = E.max() + 0.1*rg
bv = E.min() - 0.1*rg
ax.set_xlim(a.min(), a.max())
ax.set_ylim(bv, tv)

ax.set_xlabel('lattice constant', fontsize=20)
ax.set_ylabel('energy', fontsize=20)
```
