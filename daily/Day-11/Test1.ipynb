```python
import math

import matplotlib.pyplot as plt


theta = 80;

v0 = 1100;

vx = v0*math.cos(math.radians(theta))
vy = v0*math.sin(math.radians(theta))


dt = 1
g = -9.81

xp = []
yp = []

x = 0
xp.append(x)
y = 0
yp.append(y)

tm = 0

for t in range(200):

    vx += 0
    x += vx*dt
    
    vy += g*dt
    y += vy*dt

    xp.append(x)
    yp.append(y)


    if y < 0:
        break

plt.plot(xp,yp,c='b')
```
