Create a new dataframe that only contains data for the seeds that you believe will be easiest to separate. (You may need to use Google to figure out how to do this.)

```python
###ANSWER###
#15 points for writing something that works. It's worth more because it's hard to do.

df_cut = df[df['Class'].isin(["SEKER","HOROZ"])]
df_cut
```

## Part 5 (5 Points)

Split the data into a training and testing set. (Make sure to change the class values so that they're 0 and 1, instead of the actual seed names.) Also, for stupid reasons, set the `random_state` to be 35.

```python
#Write your code for splitting your data here
```

```python
###ANSWER###
#2 points for changing class to 0/1
#2 points for using train_test_split
#1 points for using random_state 35

from sklearn.model_selection import train_test_split

labels = df_cut["Class"].replace(["SEKER","HOROZ"],[0,1])
data = df_cut.drop(columns=["Class"])
x_train, x_test, y_train, y_test = train_test_split(data, labels, train_size=0.75, random_state=35)
x_train
```

## Part 6 (20 Points)

Create and train a model (either regression or SVM) for classifying the data. You can choose which type of model you want to use. **Note:** You may get multiple warnings when creating your model (warnings are *pink*). You may also get a warning about a failure to converge. It is fine to ignore these warnings. 

```python
#Write your code for training your model here
```

```python
###ANSWER###
#Logistic regression model
#20 points for doing it. There isn't a good way to break this up

import statsmodels.api as sm 

x_train_add = sm.add_constant(x_train)
x_test_add = sm.add_constant(x_test)
model_log = sm.Logit(y_train,x_train_add).fit()
print(model_log.summary())
```

```python
###ANSWER###
#SVM model
#20 points for doing it. There isn't a good way to break this up
from sklearn.svm import SVC, LinearSVC

clf = SVC(kernel='linear') 
clf.fit(x_train, y_train)
clf
```

## Part 7 (5 Points)

Evaluate your model by finding the accuracy, recall, and precision using the training data.

```python
#Write your code for determining the accuracy, recall, and precision of your model
```

```python
###ANSWER###
#Logistic regression model
#2 points for creating the predicted values
#3 points for using the predicted values to create confusion matrix

from sklearn.metrics import confusion_matrix, accuracy_score, ConfusionMatrixDisplay,classification_report

y_pred_score_log = model_log.predict(x_test_add)
y_predict_log = (y_pred_score_log>0.5).astype('int')

print("Confusion matrix: \n ",confusion_matrix(y_test, y_predict_log) )
print("\nClassification Report: \n", classification_report(y_test,y_predict_log))
```

```python
###ANSWER###
#SVM model
#2 points for creating the predicted values
#3 points for using the predicted values to create confusion matrix

from sklearn.metrics import confusion_matrix, accuracy_score, ConfusionMatrixDisplay,classification_report

y_predict_svm= clf.predict(x_test)

print("Confusion matrix: \n ",confusion_matrix(y_test, y_predict_svm) )
print("\nClassification Report: \n", classification_report(y_test,y_predict_svm))
```
