````python
### ANSWER ###
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn import metrics

CL = Raisins["Class"].unique()
Raisins['Class'] = Raisins['Class'].replace(CL[0],0)
Raisins['Class'] = Raisins['Class'].replace(CL[1],1)

X = Raisins.drop(["Class"],axis=1)
y = Raisins["Class"]

Xc = sm.add_constant(X)
train_vectors, test_vectors, train_labels, test_labels = \
    train_test_split(Xc, y, test_size=0.25, random_state=42)

logit_model = sm.Logit(train_labels, train_vectors)
result = logit_model.fit()
print(result.summary())

predicted = result.predict(test_vectors)

nominal = [1.0 if x >0.5 else 0.0 for x in predicted]
# print(nominal)
print("Accuracy:", metrics.accuracy_score(nominal, test_labels))
````

````python
import random 
import numpy as np
import math
import matplotlib.pyplot as plt
import time  
from IPython.display import display, clear_output
````

````python
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
from matplotlib.cbook import get_sample_data

def imscatter(x, y, image, ax=None, zoom=1):
    if ax is None:
        ax = plt.gca()
    try:
        image = plt.imread(image)
    except TypeError:
        # Likely already an array...
        pass
    im = OffsetImage(image, zoom=zoom)
    x, y = np.atleast_1d(x, y)
    artists = []
    for x0, y0 in zip(x, y):
        ab = AnnotationBbox(im, (x0, y0), xycoords='data', \
                            frameon=False)
        artists.append(ax.add_artist(ab))
    ax.update_datalim(np.column_stack([x, y]))
    ax.autoscale()
    return artists
````

**Human class**

````python
class human():
    
    ## initiate and parameters
    def __init__(self, roam_range, initial_x, initial_y, \
                 lower_bond, upper_bond, image_path):
        self.range = roam_range
        self.lw = lower_bond
        self.up = upper_bond
        self.image_path = image_path
        self.x = initial_x
        self.y = initial_y    

    ## draw images  
    def draw(self):
        imscatter(self.x, self.y, self.image_path, zoom=0.3)  
        
    ## random roaming  
    def roam(self):  
        self.dx = random.randint(-self.range, self.range)
        self.dy = random.randint(-self.range, self.range)
    
        # Clamp x and reverse direction if out of bounds
        if self.x <= self.lw:
            self.x = self.lw
            self.dx = abs(self.dx)
        elif self.x >= self.up:
            self.x = self.up
            self.dx = -abs(self.dx)
    
        # Clamp y and reverse direction if out of bounds
        if self.y <= self.lw:
            self.y = self.lw
            self.dy = abs(self.dy)
        elif self.y >= self.up:
            self.y = self.up
            self.dy = -abs(self.dy)
    
        self.x += self.dx
        self.y += self.dy        
````

````python
class zombie():
    
    ## initiate and parameters
    def __init__(self, roam_range, hunt_radius, kill_rate, \
                 initial_x, initial_y, lower_bond, upper_bond, image_path):
        self.range = roam_range
        self.radius = hunt_radius
        self.kr = kill_rate
        self.x = initial_x
        self.y = initial_y
        self.lw = lower_bond
        self.up = upper_bond
        self.image_path = image_path

    ## draw images    
    def draw(self):
        imscatter(self.x, self.y, self.image_path, zoom=0.3)  
        
    ## random roaming   
    def roam(self):
        self.dx = random.randint(-self.range, self.range)
        self.dy = random.randint(-self.range, self.range)
    
        # Clamp x and reverse direction if out of bounds
        if self.x <= self.lw:
            self.x = self.lw
            self.dx = abs(self.dx)
        elif self.x >= self.up:
            self.x = self.up
            self.dx = -abs(self.dx)
    
        # Clamp y and reverse direction if out of bounds
        if self.y <= self.lw:
            self.y = self.lw
            self.dy = abs(self.dy)
        elif self.y >= self.up:
            self.y = self.up
            self.dy = -abs(self.dy)
    
        self.x += self.dx
        self.y += self.dy

    ## zombie hunts human
    def hunt(self, human):
        ## part I
        self.h_x = human.x
        self.h_y = human.y

        ## Part II
        dst = math.sqrt((self.x-self.h_x)**2 + (self.y-self.h_y)**2)
        
        catch = False 
        ## Part III
        if dst < self.radius:
            if random.random() < self.kr:
                catch = True
        return catch
        
    ## Part IV    
    def infect(self):
        new_zombie = zombie(self.range, self.radius, self.kr, \
                            self.x, self.y, self.lw, self.up, \
                            self.image_path) 
        return new_zombie
      
````

````python
### ANSWER ###

## lower and upper bonds of the field
lowerB = 0
upperB = 100

## initialize zombie and human classes
n_z = 3
zombies = []
for _ in range(n_z):
    ix = random.randint(lowerB, upperB)
    iy = random.randint(lowerB, upperB)      
    wolf = zombie(2, 3, 0.85, ix, iy, lowerB, upperB, 'zombie.jpg')
    zombies.append(wolf)

humans = []
n_h = 40
for _ in range(n_h):
    ix = random.randint(lowerB, upperB)
    iy = random.randint(lowerB, upperB)   
    sheep = human(3, ix, iy, lowerB, upperB, 'human.jpg')
    humans.append(sheep)
    

new_zombies = []
z_tmp = []


## number of time iterations
nstep = 300

## solution
hpopul = []
zpopul = []
days = []
cnt = 1
## solution

## iterate 
for d in range(nstep):

    # <=== for plotting, do not change
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.set_aspect('equal', adjustable='box')
    ax.set_xlim(lowerB, upperB)
    ax.set_ylim(lowerB, upperB)   
    # ===>


    
    z_tmp = zombies
    
    survivors = []
    ## loop over human classes
    for h in range(np.size(humans)):
        hm = humans[h]     

        ## Part V ===>
        ## check whether there are zombies around
        for z in range(np.size(zombies)):
            zb = zombies[z] 
            catch = zb.hunt(hm)   
            ## zombie catch and infect human
            if catch:
                new_z = zb.infect()
                break
        
        ## if a zombie catches a human, that human turns to a zombie.
        ## otherwise survives
        if catch:
            del hm
            z_tmp.append(new_z)
        else:
            survivors.append(hm)
        ## <===    
    
    ## replace humans list with survived ones
    humans.clear
    humans = survivors 

    ## update zombies list
    zombies.clear
    zombies = z_tmp

    ## solution
    days.append(cnt)
    hpopul.append(len(humans))
    zpopul.append(len(zombies))   
    cnt += 1
    ## solution
    

    ## zombies and humans roam
    for h1 in humans:
        h1.roam()
        
    for z1 in zombies:
        z1.roam()

    # visualization
    if d%2 == 1:
        for h in humans:
            h.draw() 

        for z in zombies:
            z.draw()

    
    # <=== for plotting, do not change
    clear_output(wait=True) # Clear output for dynamic display
    display(fig)            # Reset display
    plt.close(fig)  # Prevents memory leak
    time.sleep(0.005)
    # ===>


print('humans:', len(humans), ' zombies:', len(zombies))    
````
