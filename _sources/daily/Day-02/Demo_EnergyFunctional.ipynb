### This is a simple Jupyter Notebook file for demonstrating energy functional.

Recall thermodynamics. For a mixture, the Gibbs free energy of mixing is a function of fraction (concentration): $G(X)$, where $X$ is the concentraion. Here, we recall the regular solution model of mixing: $$G(X) = M\cdot X \cdot (1-X) + X \cdot\ln(X) + (1-X)\cdot \ln(1-X),$$ where $0<X<1$. 
In a system, if we have a distribution of $X$, namely, a concentration profile: $X(r)$, where $r$ is the position, the value of Gibbs free energy at position $r$ will be $G[X(r)]$. Note that $G$ is a function of a function, $X(r)$. Thus, $G$ is called a functional. In fact, $G$ is also a function of position now. The total energy of this system is obtained by integrating over the space: $$E = \int G[X(r)] dr.$$ We can envision that $E$ will change if the distribution of $X$ varies. The state of the lowest possible $E$ will correspond to the equilibrium distribution of $X$.  Or, we can say that we can find the equilibrium $X$ at the ground state. 

The two code cells below draw two $X$ distributions and their corresponding $G$ distributions. The total energy is also calculated. We can vary $X$ to see the change of $E$. You probably won't be surprised that, if $M<2$, the lowest total energy can be achieved is when $X(r)$ is uniform. That is why diffusion occurs -- to decrease the total energy. 

```python
import numpy as np
import matplotlib.pyplot as plt

# parameters
a = 0.5
b = (1 - a) / 2
M = 1.0

# data
r = np.linspace(0, 1, 101)
X = a * r + b

# function
G = M * X * (1 - X) + X * np.log(X) + (1 - X) * np.log(1 - X)

# plots
fig, ax = plt.subplots(1, 2, figsize=(8, 3), sharex=True)

ax[0].plot(r, X, linewidth=2)
ax[0].set_xlim(0, 1); ax[0].set_ylim(-0.01, 1.01)
ax[0].grid(True)
ax[0].set_xlabel('r'); ax[0].set_ylabel('X')

ax[1].plot(r, G, linewidth=2)
ax[1].set_xlim(0, 1); ax[1].set_ylim(-0.5, 0.01)
ax[1].grid(True)
ax[1].set_xlabel('r'); ax[1].set_ylabel('G')

fig.tight_layout()
plt.show()

# trapezoidal integral
Et = np.trapz(G, r)
print("Et =", Et)
```

```python
# parameters

a = 0.0
b = 0.5 - a*2/np.pi
M = 1.0

# data
r = np.linspace(0, 1, 101)
X = a*np.sin(np.pi*r) + b

# function
G = M * X * (1 - X) + X * np.log(X) + (1 - X) * np.log(1 - X)

# plots
fig, ax = plt.subplots(1, 2, figsize=(8, 3), sharex=True)

ax[0].plot(r, X, linewidth=2)
ax[0].set_xlim(0, 1); ax[0].set_ylim(-0.01, 1.01)
ax[0].grid(True)
ax[0].set_xlabel('r'); ax[0].set_ylabel('X')

ax[1].plot(r, G, linewidth=2)
ax[1].set_xlim(0, 1); ax[1].set_ylim(-0.5, 0.01)
ax[1].grid(True)
ax[1].set_xlabel('r'); ax[1].set_ylabel('G')

fig.tight_layout()
plt.show()

# trapezoidal integral
Et = np.trapz(G, r)
print("Et =", Et)
```
